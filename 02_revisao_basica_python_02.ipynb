# Condicionais (if e else)

## Condicional if

O condicional if é utilizado para passar uma instrução condicionada ao computador, no caso da condição ser verdadeira. A estrutura do condicional if é sempre a seguinte

* if declaração:
* ------ instrução 01
* ------ instrução 02
* ------ .
* ------ .
* ------ .

```python
num = float(input('digite um número:'))

# Após a condição sempre deve haver os dois pontos :
if num > 0:
    # Todo bloco de código identado se encontra dentro do condicional if
    print(num, 'é um número positivo.')
    
# Note que este bloco não indentado está fora do condicional
num
```

```python
a = int(input("Digite um número inteiro: "))

if (a % 2) == 1:
    print("Número ímpar")
```

```python
a = int(input("Digite um número inteiro: "))

if (a % 2) == 0:
    print("Número par")
if (a % 2) == 1:
    print("Número ímpar")
```

## Condicional else

* if declaração:
* ------ instrução 01

* ------ .
* ------ .

* ------ .
* ------ .

```python
a = int(input("Digite um número inteiro: "))

if (a % 2) == 0:
    print("Número par")
else:
    print("Número ímpar")
```

```python
a = float(input("Digite o primeiro número: "))
b = float(input("Digite o segundo número: "))

if a == b:
    print("Os dois números são iguais")
else:
    # Podemos aninhar condicional if dentro de else
    if a > b:
        print("O maior número é o primeiro")
    # Podemos aninhar condicional else dentro de else
    else:
        print("O maior número é o segundo")
```

\begin{equation}
\nonumber \Delta = b^2 - 4ac

## While

```python
i = 1
while i < 5:
  print(i)
  i = i + 1
```

```python
n = int(input('Digite um número inteiro positivo:'))

i = 1
while i <= n:
    print(i)
    i = i + 1
```

```python
dividendo = int(input("Entre com o dividendo: "))
divisor = int(input("Entre com o divisor: "))

quociente = 0

while dividendo >= divisor:
    dividendo = dividendo - divisor
    quociente = quociente + 1

print("Quociente:", quociente)
print("Resto:", dividendo)
```

```python
base = float(input("Base: "))
exp = int(input("Expoente: "))

result = base
i = 1

while i < exp:
    result = result * base
    i = i + 1

print(result)
```

## Break

Vimos que o comando for percorre a lista completa. Às vezes queremos interromper a execução do comando for
antes dele percorrer a lista completa. O comando break faz com que a execução de um laço de repetição seja finalizada, passando a execução para o próximo comando após o laço. Podemos usar o comando else para executar um bloco de
comandos apenas caso o comando for ou comando while tenham sido executados sem interrupção (break).

```python
for i in range(1, 5):
    print(i)
```

```python
for i in range(1, 10):
    print(i)
    if i == 3:
        break
```

```python
for i in range(1, 10):
    if i == 3:
        break
    print(i) 
```

## Continue

O comando continue faz com que a execução atual do bloco de comandos do laço de repetição seja finalizada, passando a
execução para a próxima iteração do laço.

```python
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```

```python
# Código que imprime apenas os elementos pares da lista
for i in [2, 4, 7, 1, 0, 8, 9, 5]:
    if i % 2 == 0:
        print(i)
```

```python
# Código que imprime apenas os elementos pares da lista
# usa o continue
for i in [2, 4, 7, 1, 0, 8, 9, 5]:
    if i % 2 == 1:
        continue
    print(i)
```

```python
# Código que imprime apenas os elementos ímpares da lista
for i in [2, 4, 7, 1, 0, 8, 9, 5]:
    if i % 2 == 1:
        print(i)
```

```python
# Código que imprime apenas os elementos ímpares da lista
# usa o continue
for i in [2, 4, 7, 1, 0, 8, 9, 5]:
    if i % 2 == 0:
        continue
    print(i)
```

```python
# Imprimindo um retângulo com n linhas e m colunas (n × m caracteres *).
n = int(input("Entre com o número de linhas: "))
m = int(input("Entre com o número de colunas: "))

for i in range(n):
    print("*" * m)
```

```python
# Imprimindo um retângulo com n linhas e m colunas (n × m caracteres *).
n = int(input("Entre com o número de linhas: "))
m = int(input("Entre com o número de colunas: "))

for i in range(n):
    for j in range(m):
        print("*", end = "")
    print()
```
